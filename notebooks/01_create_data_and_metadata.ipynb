```python
import os
import glob
import numpy as np
import pandas as pd

import pydub
from pydub import AudioSegment
```

```python
os.makedirs('InfantVocAudio', exist_ok=True)
```

```python
raw_audio_dir = r'C:\Users\arunps\OneDrive\Projects\Scripts\Python\VisionInfantNet\RawData'
```

```python
infant_voc_audio_path = r'C:\Users\arunps\OneDrive\Projects\Scripts\Python\VisionInfantNet\InfantVocAudio'
```

```python
def audio_split_(from_sec, to_sec, wav_file_path, new_wav_file_path):
    t1 = from_sec * 1000
    t2 = to_sec * 1000
    new_audio = AudioSegment.from_wav(wav_file_path)
    new_audio = new_audio[t1:t2]
    new_audio.export(new_wav_file_path, format = 'wav') #Exports to a wav file in the current path.
```

```python
os.chdir(raw_audio_dir)
for txt_file in glob.glob('*.txt'):
    print(txt_file)
    file_path = os.path.join(raw_audio_dir, txt_file)
    df = pd.read_csv(file_path, sep = '\t')
    df.fillna('', inplace=True)
    df['classes'] = df.iloc[:, -2] + df.iloc[: , -1]
    Class_name = {'T': 'IDS', 'N': 'ADS', 'U': 'Unknown', 'C': 'Canonical', 'X': 'NonCanonical', 'R': 'LaughCry', 'V': 'Other'}
    df['classes'] = df['classes'].map(Class_name)
    df['classes'].fillna('Other', inplace=True)
    for index, row in df.iterrows():
        print(index)
        t1 = df.loc[index, 'Adjusted Begin Time - ss.msec']
        t2 = df.loc[index, 'Adjusted End Time - ss.msec']
        lable = df.loc[index, 'classes']
        in_path = os.path.join(file_path.split('/')[0], file_path.split('.txt')[0].split('/')[-1] + '.wav')
        out_path = os.path.join(infant_voc_audio_path, file_path.split('\\')[-1].split('.txt')[0] + '_' + str(index) + '_' + lable + '.wav')
        audio_split_(t1, t2, in_path, out_path)
```

```python
#create metadata df
vocaud = []
classes = []
for file in os.listdir(infant_voc_audio_path):
    vocaud.append(file)
    label = file.split('.wav')[0].split('_')[-1]
    if (label == 'ADS') | (label == 'IDS'):
        classes.append('Adult')
    elif label == 'LaughCry':
        classes.append('Other')
    else:
        classes.append(label)
voc_df = pd.DataFrame(vocaud, columns=['file_path'])
voc_df['label'] = classes
```
